```python
def to_tidy(df, value_name):
    tidy = (
        df.melt(id_vars=["country"],var_name="year", value_name=value_name)
        
    )
    tidy["year"] = tidy["year"].astype(int)
    return tidy

tidy_pop  = to_tidy(df_pop,  "population")
tidy_life = to_tidy(df_life, "lifeExp")
tidy_gni  = to_tidy(df_gni,  "gniPPP")
```

```python
# Check for all types of missing values in each dataframe

def missing_summary(df, name):
    print(f"--- {name} ---")
    print(df.isnull().sum())
    df.info()
    print()

missing_summary(tidy_pop, "tidy_pop")
missing_summary(tidy_life, "tidy_life")
missing_summary(tidy_gni, "tidy_gni")
```

```python
import pandas as pd

def parse_number(val):
    """
    Convert strings like '123M', '14k', '2.5B' to floats.
    If val is None/NaN or a non-string, return it unchanged.
    If parsing fails, return the original string.
    """
    # 1) If it's already null or not a string, leave it alone
    if pd.isna(val) or not isinstance(val, str):
        return val

    # 2) Strip whitespace & commas
    s = val.strip().replace(",", "")

    # 3) Parse suffixes or plain float
    try:
        if s.endswith("M"):
            return float(s[:-1]) * 1e6
        if s.endswith("k"):
            return float(s[:-1]) * 1e3
        if s.endswith("B"):
            return float(s[:-1]) * 1e9
        return float(s)
    except Exception:
        # 4) If anything goes wrong, just give back the original string
        return val
```

```python

# Apply to tidy_pop and tidy_gni
tidy_pop['population'] = tidy_pop['population'].apply(parse_number)
tidy_gni['gniPPP'] = tidy_gni['gniPPP'].apply(parse_number)
tidy_life['lifeExp'] = tidy_life['lifeExp'].apply(parse_number)
```

```python
# Forward fill missing values in tidy_pop, tidy_life, and tidy_gni by country and year order
tidy_pop = tidy_pop.sort_values(['country', 'year'])
tidy_pop['population'] = tidy_pop.groupby('country')['population'].ffill()

tidy_life = tidy_life.sort_values(['country', 'year'])
tidy_life['lifeExp'] = tidy_life.groupby('country')['lifeExp'].ffill()
tidy_life["lifeExp"] = (tidy_life.groupby("country")["lifeExp"].apply(lambda series: series.bfill()).reset_index(level=0, drop=True))

tidy_gni = tidy_gni.sort_values(['country', 'year'])
tidy_gni['gniPPP'] = tidy_gni.groupby('country')['gniPPP'].ffill()
```

```python
df_merged = (
    tidy_pop
      .merge(tidy_life, on=["country","year"])
      .merge(tidy_gni,  on=["country","year"])
)

print(df_merged.head(5))
print('-' * 20)
# Check for missing values
df_merged.info()
print('-' * 20)
# Check the shape of the merged dataframe
df_merged.describe()
df_merged.shape
```
